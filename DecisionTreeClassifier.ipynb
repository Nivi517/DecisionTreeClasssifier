```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
df=pd.read_csv('carsdetails.csv')
print(df.to_string())
```

```python
x=df.drop(columns=['vahicle'])
y=df['vahicle']
print(x)
print(y)
```

```python
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
ans =model.predict([[22,200000],[35,500000],[50,2000000],[15,300000]])
print(ans)
```

```python
import warnings
warnings.filterwarnings('ignore')
n=int(input())
l=[]
ans=[]
for i in range(n):
    l=list(map(int,input().split(',')))
    ans.append(model.predict([l]))
print(ans)
```
